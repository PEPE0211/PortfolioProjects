```python
from google.colab import drive
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv('/content/gdrive/My Drive/movies.csv')
```

```python
data.head(10)
```

```python
print(data.dtypes)
```

```python
#checking for missing data
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#dropping the rows with null data
data = data.dropna()
```

```python
print(np.mean(data.isnull()))
```

```python
#outliers
data.boxplot(column=['gross'])
```

```python
#deleting duplicates
data.drop_duplicates()
```

```python
data['budget'] = data['budget'].astype(int)
```

```python
data.head()
```

```python
data['yearcorrect'] = data['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
data.head(10)
```

```python
# Ordering the Data

data.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
sns.regplot(x="gross", y="budget", data=data)
```

```python
sns.regplot(x="score", y="gross", data=data)
```

```python
# Correlation Matrix between all numeric columns

data.corr(method ='pearson')
```

```python
data.corr(method ='kendall')
```

```python
data.corr(method ='spearman')
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
# Using factorize

data.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = data.apply(lambda x: x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
df_numerized = data.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method= 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```
